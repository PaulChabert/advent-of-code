```python
def count_xmas(line):
    count_xmas = 0
    for l in range(len(line)-3):
        token = line[l:l+4]
        if token == 'XMAS' or token  == 'SAMX':
            count_xmas += 1
    return count_xmas
```

```python
def access(grid,coor,off):
    try:
        x = coor[0]+off[0]
        y = coor[1]+off[1]
        if x >= 0 and y >= 0:
            return grid[x][y]
        else:
            return ''
    except(IndexError, TypeError):
        return ""

def corners(grid,coor):
    return access(grid,coor,[-1,-1]) + access(grid,coor,[-1,1]) + access(grid,coor,[1,1]) + access(grid,coor,[1,-1])
```

```python
def count_cross(grid):
    count_cross = 0
    for r in range(len(grid)):
        for c in range(len(grid[0])):
            if grid[r][c] == 'A':
                corner_string = corners(grid,[r,c])
                if corner_string in ('MMSS','SSMM','SMMS','MSSM'):
                    count_cross += 1
    return count_cross
```

```python
def matrix_choose(grid):
    matrix = [['-']*len(grid[0]) for _ in range(len(grid))]
    for r in range(len(grid)):
        for c in range(len(grid[0])):
            if grid[r][c] == 'A':
                corner_string = sorted(corners(grid,[r,c]))
                if corner_string == ['M','M','S','S']:
                    matrix[r][c] = 'X'

    return matrix
```

```python
# chat gpt answer

with open("input_4.txt") as f:
    grid = [line.strip() for line in f]

rows = len(grid)
cols = len(grid[0])
count = 0

# Loop over all possible center positions for the "X" shape
for i in range(1, rows - 1):
    for j in range(1, cols - 1):
        center = grid[i][j]
        if center != 'A':
            continue

        # Check diagonal from top-left to bottom-right: (i-1,j-1), (i,j), (i+1,j+1)
        d1 = grid[i-1][j-1] + center + grid[i+1][j+1]
        # Check diagonal from top-right to bottom-left: (i-1,j+1), (i,j), (i+1,j-1)
        d2 = grid[i-1][j+1] + center + grid[i+1][j-1]

        valid = {"MAS", "SAM"}  # Forward or backward

        if d1 in valid and d2 in valid:
            count += 1

print("Number of X-MAS patterns found:", count)
```
