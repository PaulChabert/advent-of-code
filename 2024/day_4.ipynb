```python
data = open('input_4.txt',"r").readlines()
data = [line.strip() for line in data]
```

```python
def count_xmas(line):
    count_xmas = 0
    for l in range(len(line)-3):
        token = line[l:l+4]
        if token == 'XMAS' or token  == 'SAMX':
            count_xmas += 1
    return count_xmas
```

```python
def vertical_matrix(data):
    return [ ''.join(chars) for chars in list(zip(*data))]
```

```python
def count_xmas_one(line):
    import re
    count_xmas = 0
    for l in range(len(line)-4):
        token = line[l:l+5]
        if bool(re.search("X.MAS",line)) or bool(re.search("X.MAS",line)):
            count_xmas += 1
    return count_xmas
```

```python
def access(grid,coor,off):
    try:
        return grid[coor[0]+off[0]][coor[1]+off[1]]
    except(IndexError, TypeError):
        return ""

def corners(grid,coor):
    return access(grid,coor,[-1,-1]) + access(grid,coor,[-1,1]) + access(grid,coor,[1,1]) + access(grid,coor,[1,-1])
```

```python
def count_cross(grid):
    count_cross = 0
    for r in range(len(grid)):
        for c in range(len(grid[0])):
            if grid[r][c] == 'A':
                corner_string = sorted(corners(grid,[r,c]))
                if corner_string == ['M','M','S','S']:
                    count_cross += 1
    return count_cross
```
