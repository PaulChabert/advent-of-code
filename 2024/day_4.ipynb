```python
data = [line.strip() for line in data]
data = [line.strip() for line in data]
```

```python
def count_xmas(line):
    count_xmas = 0
    for l in range(len(line)-3):
        token = line[l:l+4]
        if token == 'XMAS' or token  == 'SAMX':
            count_xmas += 1
    return count_xmas
```

```python
def vertical_matrix(data):
    return [ ''.join(chars) for chars in list(zip(*data))]
```

```python
def offset(lines,reverse = False):
    if reverse:
        return ['-' * (len(lines)-1-n) + line + '-'*n for n, line in enumerate(lines)]
    else: 
        return ['-'*n + line + '-' * (len(lines)-1-n) for n, line in enumerate(lines)]
```
